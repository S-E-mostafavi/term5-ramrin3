````python
# TODO: ูุชุฏ ุจุฑุง ูพุด ูพุฑุฏุงุฒุด  ูุถุง ุฐุฎุฑู ุณุงุฒ ุทุจู ุชูุถุญุงุช ุงุฑุงุฆู ุดุฏู ุจููุณุฏ.

import pandas as pd
import numpy as np

def preprocess_storage(df):
        
    # ุงุฌุงุฏ ฺฉูพ ุงุฒ ุฏุชุงูุฑู ุงุตู
    df_processed = df.copy()
    
    df_processed['Storage_Type'] = df_processed['Storage'].apply(
        lambda x: ''.join([i for i in x if not i.isdigit()]).strip()
    )
    
    def convert_to_gb(storage_str):
        num_str = ''.join([c for c in storage_str if c.isdigit() or c == '.'])
        
        if not num_str:
            return np.nan
            
        num = float(num_str)
        
        if 'TB' in storage_str.upper():
            return num * 1024
        elif 'GB' in storage_str.upper():
            return num
        else:
            # ุงฺฏุฑ ูุงุญุฏ ูุดุฎุต ูุจูุฏุ ูุฑุถ ูโฺฉูู ฺฏฺฏุงุจุงุช ุงุณุช
            return num
    
    df_processed['Storage_GB'] = df_processed['Storage'].apply(convert_to_gb)
    
    # ุฏุณุชูโุจูุฏ ูุถุง ุฐุฎุฑูโุณุงุฒ
    bins = [0, 256, 512, 1024, np.inf]
    labels = ['<256GB', '256GB-512GB', '512GB-1TB', '>1TB']
    df_processed['Storage_Category'] = pd.cut(
        df_processed['Storage_GB'],
        bins=bins,
        labels=labels,
        right=False
    )
    
    return df_processed

df = pd.DataFrame(df)
df_processed = preprocess_storage(df)
print(df_processed[['Brand', 'Storage', 'Storage_Type', 'Storage_GB', 'Storage_Category']])
````

````python
# TODO ูุชุฏ ุจุฑุง ูุฏุฑุช ูฺฺฏ ูุง GPU ู Processor ุจููุณุฏ.
def parse_gpu_processor(df):
   
    df = df.copy()
    
    # ุชุฌุฒู Processor
    df[['Processor_Brand', 'Processor_Model']] = df['Processor'].str.extract(r'([A-Za-z]+)\s*(.*)')
    
    # ุชุฌุฒู GPU
    gpu_split = df['GPU'].str.extract(r'([A-Za-z]+)\s*([A-Za-z]*)\s*(\d+)')
    df['GPU_Brand'] = gpu_split[0]
    df['GPU_Series'] = gpu_split[1]
    df['GPU_Model'] = gpu_split[2]
    
    
    df['GPU_Performance'] = df['GPU'].apply(classify_gpu_performance)
    df['CPU_Performance'] = df['Processor'].apply(classify_cpu_performance)
    
    return df

def classify_gpu_performance(gpu_str):
    """ุชุนู ุณุทุญ ุนููฺฉุฑุฏ GPU"""
    if 'RTX 3080' in gpu_str: return 'High-End'
    elif 'RTX 3060' in gpu_str: return 'Mid-High'
    elif 'GTX 1650' in gpu_str: return 'Mid'
    elif 'RX 6600' in gpu_str: return 'Mid-High'
    else: return 'Unknown'

def classify_cpu_performance(cpu_str):
    if 'Ryzen 7' in cpu_str or 'i7' in cpu_str: return 'High-End'
    elif 'Ryzen 5' in cpu_str or 'i5' in cpu_str: return 'Mid'
    elif 'Ryzen 3' in cpu_str or 'i3' in cpu_str: return 'Entry'
    else: return 'Unknown'
````

````python
# TODO: ุงุฑุฒุงุจ ูุฏู ุฑู ูุฌููุนูโูุง ูุฎุชูู ุจุง ุงุณุชูุงุฏู ุงุฒ ูุชุฏ evaluate_model

import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python


def evaluate_model(model, X, y, cv=5, test_size=0.2, random_state=42):
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    
    #  ุขููุฒุด ูุฏู
    model.fit(X_train, y_train)
    
    # ูพุดโุจู ุฑู ุฏุงุฏูโูุง ุขููุฒุด ู ุขุฒููู
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    metrics = {
        'train': {
            'MAE': mean_absolute_error(y_train, y_train_pred),
            'MSE': mean_squared_error(y_train, y_train_pred),
            'RMSE': np.sqrt(mean_squared_error(y_train, y_train_pred)),
            'R2': r2_score(y_train, y_train_pred)
        },
        'test': {
            'MAE': mean_absolute_error(y_test, y_test_pred),
            'MSE': mean_squared_error(y_test, y_test_pred),
            'RMSE': np.sqrt(mean_squared_error(y_test, y_test_pred)),
            'R2': r2_score(y_test, y_test_pred)
        }
    }
    
    cv_scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
    metrics['cross_val'] = {
        'mean_R2': np.mean(cv_scores),
        'std_R2': np.std(cv_scores),
        'all_scores': cv_scores
    }
    
    plot_results(y_test, y_test_pred, model.__class__.__name__)
    
    return metrics

def plot_results(y_true, y_pred, model_name):
    
    plt.figure(figsize=(15, 5))
    
    # ููุงุณู ููุงุฏุฑ ูุงูุน ู ูพุดโุจู ุดุฏู
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
    plt.xlabel('Actual Price ($)')
    plt.ylabel('Predicted Price ($)')
    plt.title(f'Actual vs Predicted - {model_name}')
    

model = RandomForestRegressor(n_estimators=100, random_state=42)

results = evaluate_model(model, X, y)

#  ููุงุด ูุชุงุฌ
print("\nูุชุงุฌ ุงุฑุฒุงุจ ูุฏู:")
print(f"Rยฒ Score - Train: {results['train']['R2']:.4f}")
print(f"Rยฒ Score - Test: {results['test']['R2']:.4f}")
print(f"Cross-Val Rยฒ: {results['cross_val']['mean_R2']:.4f} ยฑ {results['cross_val']['std_R2']:.4f}")
print(f"MAE: {results['test']['MAE']:.2f}")
print(f"RMSE: {results['test']['RMSE']:.2f}")
````

````python
def plot_learning_curve(model, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 5)):
   
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y, cv=cv, scoring='r2', train_sizes=train_sizes, n_jobs=-1
    )
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)
    
    # ุฑุณู ููุญู ุงุฏฺฏุฑ
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training score')
    plt.plot(train_sizes, val_mean, 'o-', color='g', label='Cross-validation score')
    
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='g')
    
    plt.xlabel('Training examples')
    plt.ylabel('Rยฒ Score')
    plt.title(f'Learning Curve - {model.__class__.__name__}')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

plot_learning_curve(model, X, y)
````
